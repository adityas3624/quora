```python
nltk.download('punkt')
```

```python
inputs= train_df['question_text'].apply(nltk.word_tokenize)
```

```python
inputs
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
english_stopwords = stopwords.words('english')
```

```python
def remove_stopwords(tokens):
    return [word for word in tokens if word not in english_stopwords]
```

```python
my_list=[]
for list in inputs:
    my_list.append(remove_stopwords(list))
```

```python
join=[' '.join(list) for list in my_list]
```

```python
nltk.download('wordnet')
```

```python
nltk.download('wordnet')
from nltk.stem import PorterStemmer

# Create a PorterStemmer instance
stemmer = PorterStemmer()
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer


def tokenize(text):
    return [stemmer.stem(word) for word in word_tokenize(text)]
vectorizer = CountVectorizer(lowercase=True, 
                             tokenizer=tokenize,
                             stop_words=english_stopwords,
                             max_features=10000000000)
```

```python
vectorizer.fit(join)
```

```python
inputs_df=vectorizer.transform(join)
```

```python
test_inputs = vectorizer.transform(test_df.question_text)
```

```python
test_inputs.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_inputs, val_inputs, train_targets, val_targets = train_test_split(inputs_df, train_df.target, 
                                                                        test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression(solver='saga')
```

```python
%%time
model.fit(train_inputs, train_targets)
```

```python
train_preds=model.predict(train_inputs)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_preds,train_targets)
```

```python
from sklearn.metrics import f1_score
f1_score(train_preds,train_targets)
```

```python
val_preds=model.predict(val_inputs)
```

```python
f1_score(val_preds,val_targets)
```

```python
model=LogisticRegression(solver='sag')
model.fit(train_inputs, train_targets)
train_preds=model.predict(train_inputs)
accuracy_score(train_preds,train_targets)
```

```python
test_preds=model.predict(test_inputs)
```

```python
vul=['racist','sex','sexist','penis','peni','arab','american','european','raci','race','dictatorship','republican']
```

```python
test_preds[0]
```

```python
for n in range(test_preds.shape[0]):
    t=word_tokenize(test_df.question_text[n])
    for word in t:
        if word in vul:
            test_preds[n]=1
```

```python
sub_df=pd.read_csv('/kaggle/input/quora-insincere-questions-classification/sample_submission.csv')
```
